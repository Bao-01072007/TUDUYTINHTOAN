### W5A16
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x



```python
# Số Armstrong (hay còn gọi là số tự khiêu - Narcissistic number) là một số mà tổng các chữ số của nó, mỗi chữ số được nâng lên lũy thừa bậc $n$ (với $n$ là số chữ số), bằng chính nó.
def is_armstrong(x:int) -> bool:
    # Chuyển số thành chuỗi để lấy số chữ số
    s = str(x)
    n = len(s)
    tong = 0
    for i in s: # Duyệt các phần tử trong chuỗi s
        chu_so = int(i) # Chuyển các phần tử thành số nguyên
        tong += chu_so**n
    if tong == x:
        return True
    else:
        return False

# Test
a = int(input("Nhập số"))
if is_armstrong(a):
    print("True")
else:
    print("False")
```

### W5A17
Viết hàm kiểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
import math
def check_so_chinh_phuong(n):
    if n < 0:
        return False
    s = int(math.sqrt(n))
    if s * s == n:
        return True
    return False
    
# Test
a = int(input())
print(check_so_chinh_phuong(a))
```

### W5A18 
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def check_so_doi_xung(x):
    dao_nguoc_so = 0
    gia_tri_ban_dau = x  # Gán giá trị ban đầu bằng x để tí đem đi so sánh
    while x > 0:
        chu_so_cuoi = x%10
        dao_nguoc_so = dao_nguoc_so * 10 + chu_so_cuoi
        x = x // 10
    return dao_nguoc_so == gia_tri_ban_dau

# Test
a = int(input())
if check_so_doi_xung(a):
    print("Là số đối xứng")
else:
    print("Không là số đối xứng")
```

### W5A19 
VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def check_so_ki_dieu(n):
    temp = n
    while temp >= 10: # Lặp đến khi nào còn 1 chữ số
        tong = 0
        while temp > 0:
            chu_so_cuoi = temp%10
            tong += chu_so_cuoi
            temp = temp//10
        temp = tong
    return temp == 1

# Test
a = int(input())
if check_so_ki_dieu(a):
    print("True")
else:
    print("False")        
    

    
```

### W5A25
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)


```python
def cross_product(a, b, p):
    """Tính tích có hướng của vectơ AB và AP"""
    # (x_B - x_A)*(y_P - y_A) - (y_B - y_A)*(x_P - x_A)
    return (b[0] - a[0]) * (p[1] - a[1]) - (b[1] - a[1]) * (p[0] - a[0])

def point_in_quadrilateral(quad, p):
    # Lấy 4 đỉnh của tứ giác
    p1, p2, p3, p4 = quad
    
    # Tính tích có hướng cho 4 cạnh: (P1,P2), (P2,P3), (P3,P4), (P4,P1)
    cp1 = cross_product(p1, p2, p)
    cp2 = cross_product(p2, p3, p)
    cp3 = cross_product(p3, p4, p)
    cp4 = cross_product(p4, p1, p)
    
    # Nếu điểm P nằm trong hoặc trên biên, tất cả tích có hướng 
    # phải cùng dấu (tất cả >= 0 hoặc tất cả <= 0)
    is_all_non_negative = (cp1 >= 0 and cp2 >= 0 and cp3 >= 0 and cp4 >= 0)
    is_all_non_positive = (cp1 <= 0 and cp2 <= 0 and cp3 <= 0 and cp4 <= 0)
    
    return is_all_non_negative or is_all_non_positive

# --- Test thử ---
quad = [(0,0), (4,0), (4,3), (0,3)]
p = (2,1)
print(f"Điểm {p} nằm trong tứ giác: {point_in_quadrilateral(quad, p)}") # True

p_edge = (4, 1.5)
print(f"Điểm {p_edge} nằm trên biên: {point_in_quadrilateral(quad, p_edge)}") # True
```

```python
def SL_chu_so_khac_0(n: list):
    dem = 0
    for i in n:
        temp = abs(i)
        while temp > 0:
            chu_so_cuoi = temp % 10
            if chu_so_cuoi != 0:
                dem += 1
            temp = temp //10
    return dem

# Test
s = list(map(int, input().split()))
print(SL_chu_so_khac_0(s))

```

### W5A28 
Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math
def tinh_goc(x1, y1, x2, y2, x3, y3):
    # Tính tọa độ vecto BA và BC
    x_ba, y_ba = x1-x2, y1-y2
    x_bc, y_bc = x3-x2, y3-y2
    tich_vo_huong = x_ba*x_bc + y_ba*y_bc
    # Tính tích độ dài
    do_dai_ba = math.sqrt(x_ba**2 + y_ba**2)
    do_dai_bc = math.sqrt(x_bc**2 + y_bc**2)
    tich_do_dai = do_dai_ba*do_dai_bc
    cosin_goc = tich_vo_huong/tich_do_dai
    # Giới hạn giá trị trong khoảng [-1, 1]
    cosin_goc = max(-1.0, min(1.0, cosin_goc))
    goc_radian = math.acos(cosin_goc)
    goc_do = math.degrees(goc_radian)
    return goc_do

# Test
print(tinh_goc(0, 1, 0, 0, 1, 0))
```

### W5A30 V
Viết hàm tính hệ số nhị thức C(n, k).


```python
import math

def nCr(n, k):
    return math.comb(n, k)

# --- Test ---
print(nCr(5, 2))  # Kết quả: 10
```

```python
def nCr_manual(n, k):
    # Các trường hợp đặc biệt
    if k < 0 or k > n:
        return 0
    if k == 0 or k == n:
        return 1
        
    # Tối ưu: C(n, k) = C(n, n-k). Ví dụ: C(10, 8) tính cực hơn C(10, 2)
    if k > n // 2:
        k = n - k
        
    ket_qua = 1
    for i in range(k):
        ket_qua = ket_qua * (n - i) // (i + 1)
        
    return ket_qua

# --- Test ---
print(nCr_manual(10, 3)) # Kết quả: 120
```

### W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.


```python
def check_so_khac_nhau(x):
    s = str(x) # Chuyển số thành chuỗi để dùng hàm set
    return len(set(s)) == len(s) # Nếu độ dài chuỗi sau khi bỏ các phần tử trùng lặp mà vẫn bằng chuỗi ban đầu thì là đôi một khác nhau
def sum_1_to_n(n):
    if n <= 0:
        return 0
    tong = 0
    for i in range(1, n+1):
        if check_so_khac_nhau(i):
            tong += i
    return tong

# Test
n = int(input())
kq = sum_1_to_n(n)
print(kq)
```
