### W4A21 
Nhập vào số nguyên dương n, tính tổng các chữ số của n 


```python
n = int(input("Nhập vào số nguyên dương n"))
n = abs(n)
tong = 0
while n>0:
    chu_so_cuoi = n%10
    tong += chu_so_cuoi
    n = n//10
print(tong)
```

### W4A22 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input())
dem_chan = 0
dem_le = 0
if n == 0:
    dem_chan = 1
while n>0:
    chu_so_cuoi = n%10
    if chu_so_cuoi % 2 == 0:
        dem_chan += 1
    else:
        dem_le += 1
    n = n//10
print(f"{dem_chan} chữ số chẵn")
print(f"{dem_le} chữ số lẻ")

    
```

### W4A24
Nhập vào A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 



```python
A = int(input())
s = 0
k = 0
while s <= A:
    k += 1
    s += 1/k
print(k)
print(f"{s:.6f}")
```

### W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
numbers = [] # Khởi tạo 1 danh sách rỗng
while True: # Vòng lặp vô hạn, nhập đến bao giờ break thì thôi
    day_so = int(input("Nhập 1 dãy số nguyên, nhập -1 để dừng"))
    if day_so == -1:
        break
    numbers.append(day_so)
if len(numbers)>0: # Tránh trường hợp nhập ngay -1 ở đầu
    print(f"Số lớn nhất: {max(numbers)}")
    print(f"Số bé nhất: {min(numbers)}")
else:
    print("Vui lòng nhập hợp lệ")
    
```

### W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 



```python
A = int(input())
f1, f2 = 1, 1 # Hai số bắt đầu của dãy là 1, 1
while True:
    n = f1+f2
    if n > A:
        break
    f1, f2 = f2, n
print(f2)

```

### W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 


```python
chuoi = input()
chu_hoa = 0
chu_thuong = 0
chu_so = 0
for i in chuoi:
    if i.isupper():
        chu_hoa += 1
    elif i.islower():
        chu_thuong += 1
    elif i.isdigit():
        chu_so += 1
print(f"{chu_hoa} chữ hoa")
print(f"{chu_thuong} chữ thường")
print(f"{chu_so} chữ số")
```
