### W3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2
hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0


```python
def kiem_tra_luy_thua_cua_2(n):
    # n > 0 để loại bỏ số 0 và số âm, vì chúng không phải là lũy thừa của 2
    # (n & (n - 1) == 0) kiểm tra tính chất bitwise
    return n>0 and (n&(n-1)==0)
num = int(input())
if kiem_tra_luy_thua_cua_2(num):
    print(f"{num} là lũy thừa của 2")
else:
    print(f"{num} không phải lũy thừa của 2")


    
```

```python
def kiem_tra_luy_thua_cua_2(n):
    if n <= 0:
        return False
    while n%2==0: # Chạy đến khi nào vẫn còn có thể chia hết cho 2
        n = n//2
    return n==1 # Chia nếu kết quả cuối cùng ra 1 thì n là lũy thừa của 2
num = int(input())
if kiem_tra_luy_thua_cua_2(num):
    print(f"{num} là lũy thừa của 2")
else:
    print(f"{num} không phải là lũy thừa của 2")
```

- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 


```python
n = float(input("Nhập vào số điện tiêu thụ"))
if 0<=n<=50:
    so_tien = n*1500
elif 51<=n<=100:
    so_tien = n*1500 + (n-50)*2000
else:
    so_tien = n*1500 +50*2000 + (n-100)*300
print(f"Số tiền phải trả là: {so_tien:.2f}") 
```

### W3A16
[Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả
cách nhau một dấu cách


```python
n = float(input())
# Nhân 10 để xử lý chữ số thập phân đầu tiên
k = int(n * 10)

# Lấy phần nguyên cơ bản
phan_nguyen = k // 10
# Lấy chữ số thập phân (hàng đơn vị của k)
chu_so_le = k % 10
if chu_so_le == 0:
    # Nếu là số nguyên (ví dụ 9.0)
    len_val = phan_nguyen
    xuong_val = phan_nguyen
    gan_nhat = phan_nguyen
else:
    # Nếu có phần lẻ (ví dụ 7.4 hoặc -5.8)
    if n > 0:
        len_val = phan_nguyen + 1
        xuong_val = phan_nguyen
        if chu_so_le >= 5:
            gan_nhat=phan_nguyen+1
        else:
            gan_nhat=phan_nguyen
    else:
        # Số âm thì logic ngược lại một chút
        len_val = phan_nguyen + 1
        xuong_val = phan_nguyen
        # Với số âm, k%10 của -5.8 là 2 (vì -58 = -6*10 + 2) 
        # nên cần cẩn thận đoạn này
        if chu_so_le >= 5:
            gan_nhat = phan_nguyen
        else:
            gan_nhat=phan_nguyen +1
print(len_val, xuong_val, gan_nhat)
```

và so sánh vs c, c nhân công bội và so sánh vs d. 

```python
a, b, c, d = map(int, input().split())
if a==0:
    if b==0 and c==0 and d==0:
        print("Tạo thành cấp số nhân")
    else:
        print("Không tạo thành cấp số nhân")
else:
    if b%a!=0:
        print("Không tạo thành cấp số nhân")
    else:
        q = b/a
        if (a*q == b) and (b*q==c) and (c*q==d):
            print("Tạo thành cấp số nhân")
        else:
            print("Không tạo thành cấp số nhân")    
```
