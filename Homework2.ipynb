### Bài 1: 
Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó. 

```python
a, b = map(int, input().split())
s = a*b
c = (a+b)*2
print(f"Diện tích là {s}")
print(f"Chu vi là {c}")
```

### Bài 2: 
Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó. 

```python
import math
r = float(input("Nhập bán kính"))
s = math.pi*(r**2)
c = 2*math.pi*r
print(f"Diện tích là: {s:.2f}")
print(f"Chu vi là: {c:.2f}")
```

### Bài 3 : 
Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
import math
a, b, c = map(float, input().split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        loai = "Tam giác đều"
    elif a==b or b==c or c==a:
        if (math.isclose(a**2 + b**2, c**2) or
            math.isclose(b**2 + c**2, a**2) or
            math.isclose(a**2 + c**2, b**2)):
            loai = "Tam giác vuông cân"
        else:
            loai = "Tam giác cân"
    elif (math.isclose(a**2 + b**2, c**2) or
          math.isclose(b**2 + c**2, a**2) or
          math.isclose(a**2 + c**2, b**2)):
          loai = "Tam giác vuông"
    else:
        loai = "Tam giác thường"
    print(loai)   
    chuvi = a + b + c
    p = chuvi/2
    s = math.sqrt(p*(p-a)*(p-b)*(p-c))   
    print(f"Chu vi là: {chuvi:.2f}")
    print(f"Diện tích là: {s:.2f}")     
else:
    print("Không là ba cạnh của một tam giác")
        
```

### Bài 5: 
Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a = float(input("Nhập hệ số a"))
b = float(input("Nhập hệ số b"))
c = float(input("Nhập hệ số c"))
denta = (b**2)-4*a*c
if denta > 0:
    x1 = (-b+math.sqrt(denta))/(2*a)
    x2 = (-b-math.sqrt(denta))/(2*a)
    print(f"Phương trình có hai nghiệm phân biệt: x1 = {x1:.2f}, x2 = {x2:.2f}")
elif denta == 0:
    x = (-b)/(2*a)
    print(f"Phương trình có nghiệm kép x1 = x2 = {x:.2f}")
else:
    print("Phương trình vô nghiệm")
```

### Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n


```python
# Dùng định thức Cramer
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
m = float(input("Nhập m: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
n = float(input("Nhập n: "))
D = a*d - b*c
Dx = d*m - b*n
Dy = a*n - c*m
if D !=0:
    x = Dx/D
    y = Dy/D
    print(f"Hệ phương trình có nghiệm duy nhất: x = {x:.2f}, y = {y:.2f}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình có vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
```

### Bài 10: 
Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào mộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math
xc = float(input("Nhập hoành độ tâm C"))
yc = float(input("Nhập tung độ tâm C"))
r = float(input("Nhập bán kính"))
xA = float(input("Nhập hoành độ điểm A"))
yA = float(input("Nhập tung độ điểm A"))
d = math.sqrt((xA-xc)**2 + (yA-yc)**2)
if math.isclose(d, r): # Sử dụng isclose để tránh sai số trên số thực
    print("Điểm A nằm trên đường tròn")
elif d < r:
    print("Điểm A nằm trong đường tròn")
else:
    print("Điểm A nằm ngoài đường tròn")
```
