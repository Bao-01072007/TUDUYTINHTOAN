### W5A3 
Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 


```python
import math
def kiem_tra_so_nguyen_to(n):
    for i in range(2, int(math.sqrt(n))+1):
        if n%i == 0:
            return False
    return True
# Ví dụ sử dụng
n = int(input())
if kiem_tra_so_nguyen_to(n):
    print("Là số nguyên tố")
else: 
    print("Không là số nguyên tố")
```

### W5A4 
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
# Số hoàn hảo (Perfect Number) là số nguyên dương mà tổng của tất cả các ước số thực sự của nó (không bao gồm chính nó) bằng chính nó.
import math
def so_hoan_hao(n):
    tong_uoc = 1 # Gán bằng 1 vì 1 là ước của mọi số nguyên dương
    for i in range(2, int(math.sqrt(n))+1):
        if n%i == 0:
            if i * i == n: # Số chính phương
                tong_uoc += 1 # Chỉ cộng 1 lần tránh lặp số chính phương
            else:
                tong_uoc += i + (n//i)
    return tong_uoc == n # Kiểm tra tổng các ước có bằng số ban đầu hay không

# Test
test_nums = [6, 28, 496, 8128, 10] # Ktra các số trong danh sách
for num in test_nums:
    if so_hoan_hao(num):
        print("Số hoàn hảo")
    else:
        print("Không là số hoàn hảo")

                
```

### W5A7 
Viết hàm giả lập một máy tính bỏ túi.
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
\)
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
Phép trừ \( - \)
Phép cộng \( + \)
Phép chia \( / \)
Phép nhân \( * \)
Kết quả làm tròn đến chữ số thập phân thứ hai.
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
Đầu ra: Kết quả phép toán

```python
def tinh_toan(so1: float, so2: float, operat: str):
    ket_qua = 0.0
    if operat == '+':
        ket_qua = so1 + so2
    elif operat == '-':
        ket_qua = so1-so2
    elif  operat == '*':
        ket_qua = so1*so2
    elif  operat == '/':
        if  so2 == 0:
            return "Không thể chia 0"
        ket_qua = so1/so2
    else:
        return "Toán tử không hợp lệ"
    return round(ket_qua, 2) # làm tròn đến 2 chữ số thập phân

# Test
a, b = map(float, input().split())
phep_tinh = input("Nhập toán tử")
dap_an = tinh_toan(a, b, phep_tinh)
print(dap_an)
```

### W5A8
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 


```python
def hamming_distance(x: int, y: int):
    xor_result = x^y # XOR 2 số để tìm vị trí bit khác nhau, giống nhau ra 0, khác nhau ra 1, vậy số lượng vị trí khác nhau là đếm xem có bao nhiêu phần tử 1
    return bin(xor_result).count('1') # Chuỗi sang chuỗi nhị phân và đếm số phần tử '1' trong chuỗi
 
# Test
a, b = map(int, input().split())
print(hamming_distance(a, b))
```

### W5A9 
Viết hàm trả về tổng các chữ số của số nguyên dương n 


```python
def tinh_cac_chu_so(n):
    n = abs(n)
    tong = 0
    while n>0:
        chu_so_cuoi = n%10
        tong += chu_so_cuoi
        n = n//10
    return tong

# Test
n = int(input())
ket_qua = tinh_cac_chu_so(n)
print(ket_qua)
```

từ đầu vào có cùng độ dài.

0 <= |r| < |divisor|.

### W5A14 
Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op

```python
def compute(a, b, op):
    ket_qua = 0.0
    if op == '+':
        ket_qua = a + b
    elif op == '-':
       ket_qua = a - b
    elif op == '*':
        ket_qua = a * b
    elif op == '/':
        if b == 0:
            return "Lỗi"
        ket_qua = a/b
    else:
        return "Không hợp lệ"
    return round(ket_qua, 4)

## Test
a, b = map(float, input().split())
op = input("Nhập toán tử")
dap_an = compute(a, b, op)
print(dap_an)
```

### W5A15 
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
import math
def is_right_triangle(a: float, b: float ,c: float):
    s = sorted([a, b, c]) # Sắp xếp theo thứ tự tăng dần
    if math.isclose(s[0]**2 + s[1]** 2, s[2]**2):
        return True
    return False

# Test 
a, b, c= map(float, input().split())
check = is_right_triangle(a,b,c)
print(check)
```
