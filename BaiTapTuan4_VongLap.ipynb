### W4A1
[SumOfNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để
tính tống các số nguyên từ 1 đến n. (n <= 1000) 


```python
n = int(input("Nhập n <= 1000"))
if n > 1000:
    print("Vui lòng nhập n<=1000")
else:
    tong = 0
    for i in range (1, n+1):  # Vòng lặp chạy từ 1 đến n
        tong += i
    print(f"Tổng các số từ 1 đến {n} là: {tong}")
```

### W4A2
[PrimeNumber]
VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi.
Kiểm tra số này có phải là số nguyên tố hay không

```python
import math
while True: # Đây là 1 vòng lặp vô hạn, nó chỉ thoát ra khi điều kiện đúng n > 0
    n = int(input("Nhập 1 số nguyên dương"))
    if n > 0:
        break
    print("Không phải số dương, vui lòng nhập lại")
so_nguyen_to = True
if n < 2:
    so_nguyen_to = False # Để loại bỏ trường hợp số 1 là số nguyên tố
else:
    # Để kiểm tra số nguyên tố có thể kiểm tra từ 2 đến n-1 hoặc chỉ cần từ 2 đến căn bậc hai của n thôi để tối ưu hóa
    for i in range (2, int(math.sqrt(n))+1):
        if n % i == 0:
            so_nguyen_to = False
            break # Thoát ra khỏi vòng lặp ngay lập tức để không cần phải kiểm tra các trường hợp khác nữa (VD: n = 8, khi ta kiểm tra thấy n chia hết cho 2 thì dừng vòng lặp luôn, không cần kiểm tra 3,4,5,6,7 nữa)
if so_nguyen_to:
    print(f"{n} là số nguyên tố")
else:
    print(f"{n} không phải số nguyên tố") 

```

### W4A8
ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 


```python
n =int(input())
def dao_nguoc_so(n): # Đảo ngược các chữ số
    so_dao = 0
    while n>0: # Lặp các lệnh bên dưới đến chừng nào vẫn còn chữ số
        chu_so_cuoi = n%10 # Lấy chữ số cuối cùng
        so_dao = so_dao*10+chu_so_cuoi # Đưa chữ số cuối cùng đó vào số đảo ngược
        n=n//10 # Loại bỏ chữ số cuối vừa lấy ra
    return so_dao
def doi_xung(n): # Kiểm tra đối xứng bằng cách so sánh với số đảo ngược
    if n==(dao_nguoc_so(n)):
        return True
    return False
step = 0
while not doi_xung(n): # Chừng nào vẫn chưa là số đối xứng thì vẫn tiếp tục lặp các lệnh bên dưới
    n = n+dao_nguoc_so(n)
    step +=1
print(f"{step} {n}")


        

```

### W4A9
[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).


```python
def check_chu_so_khac_nhau(n):
    if n<10: return True # Các số 1 chữ số thì luôn đúng
    da_xuat_hien = []  # Tạo danh sách các chữ số đã gặp 
    while n>0: # Lặp các lệnh bên dưới đến khi n vẫn lớn hơn 0
        chu_so = n%10
        if chu_so in da_xuat_hien:
            return False # Nếu thấy chữ số đấy đã có trong list kia thì dừng luôn
        da_xuat_hien.append(chu_so) # nếu chưa xuất hiện thì thêm chữ số đó vào list
        n = n//10 # Loại bỏ chữ số vừa xét
    return True
n = int(input("Nhập n"))
found = False
i = 0
while i*i<=n:
    chinh_phuong = i*i
    if check_chu_so_khac_nhau(chinh_phuong):
        print(chinh_phuong, end=" ")
        found = True
    i = i+1
if not found:
    print("No number")
```

### W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 



```python
# Dãy Collatz: 
# Khi số đang xét là số chẵn thì số tiếp theo sẽ là: n/2
# Khi số đang xét là số lẻ thì số tiếp theo là 3n+1
# Dãy số sẽ kết thúc khi chạm đến số 1
def tinh_do_dai_day_Collatz(x):
    do_dai=1 # Vì tính cả số đầu vào
    while x!=1:
        if x % 2 ==0: #Kiểm tra số chẵn
            x=x//2
        else:
            x = 3*x+1
        do_dai+=1
    return do_dai
n = int(input("Nhập số nguyên dương n"))
do_dai_max = 0
x_tuong_ung = 0 # x tương ứng với độ dài dãy lớn nhất
for x in range(1, n+1):
    do_dai_moi = tinh_do_dai_day_Collatz(x)
    if do_dai_moi > do_dai_max: # Dấu lớn hơn để tìm được x nhỏ hơn nếu có 2 số cùng độ dài dãy       
        do_dai_max = do_dai_moi
        x_tuong_ung = x
print(f"{x_tuong_ung} {do_dai_max}")

```

[ChickenAndDog]
"Vừa gà vừa chó,
Bó lại cho tròn,
Ba mươi sáu con,
Một trăm chân chẵn"
Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
số lượng chó. Ngược lại, in ra "invalid".


```python
tong_con = int(input("Nhập tổng số gà và chó"))
tong_chan = int(input("Nhập tổng số chân"))
found = False
for so_ga in range(tong_con +1): # Chạy từ 0 vì có thể 100% con chó
    so_cho = tong_con - so_ga
    if (2*so_ga + 4*so_cho)== tong_chan:
        print(f"{so_ga} {so_cho}")
        found = True
        break
if not found:
        print("invalid")
```
