### Bài 5
Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 


```python
c = input("Nhập 1 chữ cái in hoa")
if 'B'<=c<='Z':
    ma_chu = ord(c)
    ma_moi = ma_chu + 31
    s = chr(ma_moi)
    print(f"Chữ cái thường liền trước chữ cái {c} thường là: {s}")
elif c == 'A':
    ket_qua = 'z' # Chữ liền trước a là z
    print(f"Chữ cái thường liền trước chữ cái {c} thường là: {ket_qua}")
elif 'a'<=c<='z':
    print("Vui lòng nhập chữ cái in hoa")
else:
    print("Vui lòng nhập 1 chữ cái in hoa")
```

### Bài 6
Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 



```python
import math
a, b, c = map(float, input().split())
if a+b>c and a+c>b and b+c>a: # Muốn là tam giác phải thỏa mãn đồng thời 3 TH này
    n = (a+b+c)/2
    heron = n*(n-a)*(n-b)*(n-c)
    s = math.sqrt(heron)
    print(f"Diện tích tam giác là: {s:.1f}")
else:
    print("Không phải 3 cạnh của tam giác")
```

### Bài 8
Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh 


```python
ten_chu_ho = input("Tên chủ hộ:")
chi_so_cu = int(input("Chỉ số tháng trước:"))
chi_so_moi = int(input("Chỉ số tháng này:"))
so_dien = chi_so_moi - chi_so_cu
tien_chua_vat = 0
if 0 <= so_dien <= 50:
    tien_chua_vat = so_dien*1984
elif 51 <= so_dien <= 100:
    tien_chua_vat = 50*1984 + (so_dien-50)*2050
elif 101 <= so_dien <= 200:
    tien_chua_vat = 50*1984 + 50*2050 + (so_dien-100)*2380
elif 201<= so_dien <=300:
    tien_chua_vat = 50*1984 + 50*2050 + 100*2380 + (so_dien-200)*2998
elif 301<= so_dien <=400:
    tien_chua_vat = 50*1984 + 50*2050 + 100*2380 + 100*2998 + (so_dien-300)*3350
else:
    tien_chua_vat = 50*1984 + 50*2050 + 100*2380 + 100*2998 + 100*3350 + (so_dien-400)*3460
tong_tien = tien_chua_vat*1.08
tien_phai_tra = round(tong_tien)
print(f"Họ và tên: {ten_chu_ho}")
print(f"Tiền phải trả là: {tien_phai_tra}")
```

### Bài 9
Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case


```python
try:
    a = int(input("Nhập 1 số nguyên từ 0 đến 9"))
    match a:
        case 0: # Tương tự như so sánh bằng ==
            print("Không")
        case 1:
            print("Một")
        case 2:
            print("Hai")
        case 3:
            print("Ba")
        case 4:
            print("Bốn")
        case 5:
            print("Năm")
        case 6:
            print("Sáu")
        case 7:
            print("Bảy")
        case 8:
            print("Tám")
        case 9:
            print("Chín")
        case _: #Tương tự else (Các trường hợp ngoài phạm vi)
            print("Chỉ nhập số nguyên trong khoảng từ 0 đến 9")
except ValueError:
    print("Vui lòng nhập số nguyên")

            
```
